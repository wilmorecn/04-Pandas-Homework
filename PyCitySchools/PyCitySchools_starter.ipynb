# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school. 
---

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_data_complete.head()
```

```python
#calculate new columns and values
TotalSchools = school_data_complete['school_name'].value_counts().count()
StuCount = school_data_complete['Student ID'].count()
TotalBudget = school_data_complete['budget'].unique().sum()
Math_AvgScore = school_data_complete['math_score'].mean()
Reading_AvgScore = school_data_complete['reading_score'].mean()
Overall_Pass_Rate = ((Math_AvgScore + Reading_AvgScore)/2)
Math_PercentPass = school_data_complete.loc[school_data_complete['math_score']>69,:].count()['math_score']
Math_PercentPass = (Math_PercentPass/StuCount)*100
Reading_PercentPass = school_data_complete.loc[school_data_complete['reading_score']>69,:].count()['reading_score']
Reading_PercentPass = (Reading_PercentPass/StuCount)*100
```

```python
#create new DataFrame
DistrictSummary = pd.DataFrame({"Total Schools": [TotalSchools],
                           "Total Students": [StuCount],"Total Budget": [TotalBudget],
                           "Avg Math Score": [Math_AvgScore], "% Passing Math": [Math_PercentPass],
                           "Avg Reading Score": [Reading_AvgScore], "% Passing Reading": [Reading_PercentPass],
                           "Overall Pass Rate": [Overall_Pass_Rate]})

DistrictSummary
```

```python
#add formatting to columns
DistrictSummary["Total Students"] = DistrictSummary["Total Students"].map("{:,}".format)
DistrictSummary["Total Budget"] = DistrictSummary["Total Budget"].map("${:,.2f}".format)

DistrictSummary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
temp_dict_list = []
for school in school_data_complete["school_name"].unique().tolist():
    temp_dict={"School Name": school_data_complete.loc[school_data_complete["school_name"]==school]["school_name"].max(),
               "School Type": school_data_complete.loc[school_data_complete["school_name"]==school]["type"].max(), 
               "Total Students": len(school_data_complete.loc[school_data_complete["school_name"]==school]),
               "Total Budget": school_data_complete.loc[school_data_complete["school_name"]==school]["budget"].max(),
               "Per Stu Budget": (school_data_complete.loc[school_data_complete["school_name"]==school]["budget"].max())/(len(school_data_complete.loc[school_data_complete["school_name"]==school])),
               "Math Avg Score": school_data_complete.loc[school_data_complete["school_name"]==school]["math_score"].mean(),
               "Reading Avg Score": school_data_complete.loc[school_data_complete["school_name"]==school]["reading_score"].mean(),
               "Math Pct Pass": ((len(school_data_complete.loc[(school_data_complete["school_name"]==school) & (school_data_complete["math_score"]>69)]))/ (len(school_data_complete.loc[school_data_complete["school_name"]==school])))*100,
               "Reading Pct Pass": ((len(school_data_complete.loc[(school_data_complete["school_name"]==school) & (school_data_complete["reading_score"]>69)]))/ (len(school_data_complete.loc[school_data_complete["school_name"]==school])))*100,
               "Overall Pass Rate": ((((len(school_data_complete.loc[(school_data_complete["school_name"]==school) & (school_data_complete["math_score"]>69)]))/ (len(school_data_complete.loc[school_data_complete["school_name"]==school])))*100 + ((len(school_data_complete.loc[(school_data_complete["school_name"]==school) & (school_data_complete["reading_score"]>69)]))/(len(school_data_complete.loc[school_data_complete["school_name"]==school])))*100)/2)}
    temp_dict_list.append(temp_dict)
SchoolSummary = pd.DataFrame(temp_dict_list)
SchoolSummary
```

```python
#add formatting to columns
SchoolSummary["Total Students"] = SchoolSummary["Total Students"].map("{:,}".format)
SchoolSummary["Total Budget"] = SchoolSummary["Total Budget"].map("${:,.2f}".format)
SchoolSummary["Per Stu Budget"] = SchoolSummary["Per Stu Budget"].map("${:.2f}".format)

SchoolSummary
```
